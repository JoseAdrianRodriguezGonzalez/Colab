# There's a school who is investigating the relation with their Perfomance Index and other variables such as:
* Previous scores
* Extracurricular activities
* Sleep hours
* Sample question Papers practiced
* Perfomance Index
>At the moment, we know the unique variable which is different to a nummeric value it is Extracurricular Activities, so the analysis will turn with a little connotation,
>>Firstly, we are going to visualize the data,




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
url="https://raw.githubusercontent.com/JoseAdrianRodriguezGonzalez/Colab/main/probability/Student_Performance.csv"
df=pd.read_csv(url)
print(df.head())
Hours=df['Hours Studied']
Previous_Score=df['Previous Scores']
Extracurricular=df['Extracurricular Activities']
Sleep=df['Sleep Hours']
Perfomance=df['Performance Index']
SampleQuestion=df['Sample Question Papers Practiced']
colors = np.random.rand(10000)
Axes=plt.figure(figsize=(15,15))

ax1=plt.subplot(3,2,1)
plt.scatter(Hours,Perfomance,c=colors)
ax1.set_title("Hours studied/Performance")
ax1.set_xlabel("Hours Studied")
ax1.set_ylabel("Performance Index")
ax1=plt.subplot(3,2,2)
plt.scatter(Previous_Score,Perfomance,c=colors)
ax1.set_title("Previous Score/Performance")
ax1.set_xlabel("Previous Score")
ax1.set_ylabel("Performance Index")
ax1=plt.subplot(3,2,3)
plt.scatter(Sleep,Perfomance,c=colors)
ax1.set_title("Sleep Hours/Performance")
ax1.set_xlabel("Sleep hours")
ax1.set_ylabel("Performance Index")
Yes=Extracurricular[Extracurricular=='Yes']
No=Extracurricular[Extracurricular=='No']
Yes_prom=Perfomance[Yes.index]
No_prom=Perfomance[No.index]
names=['Yes','No']
datax=[len(Yes),len(No)]
ax1=plt.subplot(3,2,4)
plt.bar(names,datax)
ax1.set_title("Extracurricular activities Histogram")
ax1.set_xlabel("Values between yes/no")
ax1.set_ylabel("Quantity")
ax1=plt.subplot(3,2,5)
plt.scatter(Extracurricular,Perfomance,c=colors)
ax1.set_title("Extracurricular activities/Performance")
ax1.set_xlabel("Extracurricular activities")
ax1.set_ylabel("Perfomance Index")
ax1=plt.subplot(3,2,6)
plt.scatter(SampleQuestion,Perfomance,c=colors)
ax1.set_title("Sample Questions Practiced/Performance")
ax1.set_xlabel("Sample Questions practiced")
ax1.set_ylabel("Perfomance Index")
print(f'The covariance betweeen hours studied and Performance is {np.cov(Hours,Perfomance)[0][1]}\n')
print(f'The covariance between the Previous Score and the Performance is {np.cov(Previous_Score,Perfomance)[0][1]}\n')
print(f'The covariance between the Sleep hours and the performance is {np.cov(Sleep,Perfomance)[0][1]}\n')
print(f'The covariance between the Sample Questions Papers Practiced and the performance is {np.cov(SampleQuestion,Perfomance)[0][1]}\n')
print(f'The average of students who are doing an Extracurricular activity is {statistics.mean(Yes_prom)}')
print(f'\nThe average of students who are not doing an Extracurricular activity is {statistics.mean(No_prom)}')
print(f'Checking Correlations \n Hours Studied {np.corrcoef(Hours, Perfomance)[0][1]}')
print(f'Checking Correlations \n Previous score {np.corrcoef(Previous_Score, Perfomance)[0][1]}')
print(f'Checking Correlations \n Sleep Hours {np.corrcoef(Sleep, Perfomance)[0][1]}')
print(f'AChecking Correlations \n Practiced Exercises {np.corrcoef(SampleQuestion, Perfomance)[0][1]}')
df1 = pd.get_dummies(Extracurricular)
df = pd.concat([df, df1], axis=1).reindex(df.index)
df.drop('Extracurricular Activities', axis=1, inplace=True)

print(f'Checking Correlations \n Students who are on an Extracurricular Activity {np.corrcoef(df["Yes"], Perfomance)[0][1]}')
print(f'Analizando correlaciones\n Students who are not on an Extracurricular Activity {np.corrcoef(df["No"], Perfomance)[0][1]}')
##para entender la relacion entrer una variable binaria y numerica es por trers hipotesiss, determinar si es igual, en ese caso, no hay relaci√≥n,y la otra es observar cual es mayor
```
