# Useful functions

```python
def change(arr,index_a,index_b):
  aux=arr[index_a] #just a little change on the variables
  arr[index_a]=arr[index_b]
  arr[index_b]=aux
  return arr
def tailWeight(mean,median):
  if(median>mean):
    print("Negative Skewed, with a difference of: ",median-mean)
  elif(median<mean):
    print("Positive Skewed with a diference of: ",median-mean)
```

## Also, we are going to check the categorical variables
Sometimes, to deal with categorical data, you'll need to make some process to change the name into a numeric symbol to represent the value

First, find out the unique categorical values

```python
HQ = Incomes['Highest_Qualified_Member']
print(HQ.unique())
```

# Then:
Create the ordinal encoder

```python
#Here it'll be extracted the array of HQ.unique, that they are the categories
categories=HQ.unique()
#Change some values
change(categories,0,1)
change(categories,3,4)
#Encode the data
coder= OrdinalEncoder(categories=[categories])
#Create the data frame
data = {'Highest_Qualified_Member': categories}
df = pd.DataFrame(data)

#Create a new column called Encoded_HQ and assigns each value to ehac cell form Highest_Qualified_Member
df['Encoded_HQ'] = coder.fit_transform(df[['Highest_Qualified_Member']])
dictionary=df.to_dict() #Create a dictionary
#Mix the values
new_dictionary = {
    'Highest_Qualified_Member': dict(zip(dictionary['Highest_Qualified_Member'].values(), dictionary['Encoded_HQ'].values()))
}
Incomes['Highest_Qualified_Member'].replace(to_replace=new_dictionary['Highest_Qualified_Member'],inplace=True)
print("Moda: ",Incomes['Highest_Qualified_Member'].mode(),"Median: ",Incomes['Highest_Qualified_Member'].median())
```

```python
scaler = preprocessing.MinMaxScaler()
normalized_incomesMonth = scaler.fit_transform(Incomes[['Mthly_HH_Income']])

# Convertir los datos normalizados a un DataFrame
df_normalized = pd.DataFrame(normalized_incomesMonth, columns=['Normalized_Income'])

# Graficar el histograma usando seaborn
sns.histplot(data=df_normalized, x='Normalized_Income', kde=True)

# Calcular la media y la mediana de los datos normalizados
print("Media:", df_normalized['Normalized_Income'].mean())
print("Mediana:", df_normalized['Normalized_Income'].median())
```

```python
normalized_incomesMonth= preprocessing.normalize([Incomes['Mthly_HH_Income']])
df_normalized = pd.DataFrame(normalized_incomesMonth.T, columns=['Normalized_Income'])
sns.histplot(data=df_normalized,x='Normalized_Income',kde=True)
print(df_normalized['Normalized_Income'].mean(),df_normalized['Normalized_Income'].median())
```

>As we've seen there still exist a lot of skewness, so it'll be useful to use other transformations

```python
log_incomesMonth= np.log(Incomes['Mthly_HH_Income'])
df_normalized = pd.DataFrame({'Normalized_Income': log_incomesMonth})
sns.histplot(data=df_normalized,x='Normalized_Income',kde=True)
print(df_normalized['Normalized_Income'].mean(),df_normalized['Normalized_Income'].median())
```

```python
log_expenseMonth= np.log(Incomes['Mthly_HH_Expense'])
df_normalized = pd.DataFrame({'Normalized_expense': log_expenseMonth})
sns.histplot(data=df_normalized,x='Normalized_expense',kde=True)
print(df_normalized['Normalized_expense'].mean(),df_normalized['Normalized_expense'].median())
```

```python
log_No_of_fly= (Incomes['No_of_Fly_Members'])**(1/3)
df_normalized = pd.DataFrame({'No_of_Fly_Members': log_No_of_fly})
sns.histplot(data=df_normalized,x='No_of_Fly_Members',kde=True)
print(df_normalized['No_of_Fly_Members'].mean(),df_normalized['No_of_Fly_Members'].median())
print(df_normalized['No_of_Fly_Members'].median()-df_normalized['No_of_Fly_Members'].mean())
```

```python
log_Annual_Incomes= np.log(Incomes['Annual_HH_Income'])
df_normalized = pd.DataFrame({'Annual_HH_Income': log_Annual_Incomes})
sns.histplot(data=df_normalized,x='Annual_HH_Income',kde=True)
print(df_normalized['Annual_HH_Income'].mean(),df_normalized['Annual_HH_Income'].median())
print(df_normalized['Annual_HH_Income'].median()-df_normalized['Annual_HH_Income'].mean())
```

```python
sqrt_earning= (Incomes['No_of_Earning_Members'])**(1/3)
df_normalized = pd.DataFrame({'No_of_Earning_Members': sqrt_earning})
sns.histplot(data=df_normalized,x='No_of_Earning_Members',kde=True)
print(df_normalized['No_of_Earning_Members'].mean(),df_normalized['No_of_Earning_Members'].median())
print(df_normalized['No_of_Earning_Members'].median()-df_normalized['No_of_Earning_Members'].mean())
```
